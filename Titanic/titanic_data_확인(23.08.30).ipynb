```python
train_df = pd.read_csv('../Downloads/titanic/train.csv')
train_df
```

```python
test_df = pd.read_csv('../Downloads/titanic/test.csv')
test_df
```

```python
test_df_a = pd.merge(test_df, gdf, on='PassengerId', how='inner')
test_df_a
```

```python
df = pd.concat([train_df, test_df_a])
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
def SBYF(x):
    if (x.SibSp > 0) | (x.Parch>0):
        return 1
    else:
        return 0
df['sbyf'] = df.apply(SBYF, axis=1)
df
```

```python
def AllSP(x):
    al = x.SibSp + x.Parch
    return al
df['allsp'] = df.apply(AllSP, axis=1)
df
```
