# Classification

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
#1번 수행
train = final_train
test = Xtest

xtrain=train.drop(columns=['Survived'])
ytrain=train['Survived']

xtrain = pd.get_dummies(xtrain, columns=['Embarked', 'Pclass', 'Sex'])
model_Survived1 = RandomForestClassifier(n_estimators=200, random_state=7, criterion='log_loss', max_features='sqrt')
```

```python
model_Survived1.fit(xtrain, ytrain)

xtest = test.drop(columns=['Survived'])
xtest = pd.get_dummies(xtest, columns=['Embarked', 'Pclass', 'Sex'])
ytest = test['Survived']
y_pred = model_Survived1.predict(xtest)
confusion = confusion_matrix(ytest, y_pred)
print(confusion)
accuracy= accuracy_score(ytest, y_pred)
accuracy
```

```python
report = classification_report(ytest, y_pred)
print(report)
```

```python
#1번 수행
train = final_train
test = Xtest

xtrain=train.drop(columns=['Survived'])
ytrain=train['Survived']

xtrain = pd.get_dummies(xtrain, columns=['Embarked', 'Pclass', 'Sex'])
model_Survived = RandomForestClassifier(n_estimators=50, random_state=3, criterion='log_loss', max_features='sqrt', max_depth=6, min_samples_leaf=18, min_samples_split=8)
```

```python
model_Survived.fit(xtrain, ytrain)

xtest = test.drop(columns=['Survived'])
xtest = pd.get_dummies(xtest, columns=['Embarked', 'Pclass', 'Sex'])
ytest = test['Survived']
y_pred = model_Survived.predict(xtest)
confusion = confusion_matrix(ytest, y_pred)
print(confusion)
accuracy= accuracy_score(ytest, y_pred)
accuracy
```

```python
report = classification_report(ytest, y_pred)
print(report)
```

```python
#k번(10번) 수행
kf = KFold(n_splits=10, shuffle=True, random_state=5)
for train_index, test_index in kf.split(xtrain):
    X_train_fold, X_test_fold = xtrain.iloc[train_index], xtrain.iloc[test_index]
    y_train_fold, y_test_fold = ytrain.iloc[train_index], ytrain.iloc[test_index]

    model_Survived.fit(X_train_fold, y_train_fold)
    y_pred_fold = model_Survived.predict(X_test_fold)
    
    fold_accuracy = accuracy_score(y_test_fold, y_pred_fold)
    print(fold_accuracy)
    
    confusion = confusion_matrix(y_test_fold, y_pred_fold)
    print(confusion)

    classification_rep = classification_report(y_test_fold, y_pred_fold)
    print(classification_rep)
```

```python
from sklearn.model_selection import GridSearchCV

params = {'n_estimators' : [10, 50, 100, 150, 200],
          'random_state' : [3, 5, 7],
          'max_features' : ['sqrt', 'log2'],
          'criterion' : ['gini', 'log_loss']
            }

# RandomForestClassifier 객체 생성 후 GridSearchCV 수행
rf_clf = RandomForestClassifier(random_state = 0, n_jobs = -1)
grid_cv = GridSearchCV(rf_clf, param_grid = params, cv = 3, n_jobs = -1)
grid_cv.fit(xtrain, ytrain)

print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```
